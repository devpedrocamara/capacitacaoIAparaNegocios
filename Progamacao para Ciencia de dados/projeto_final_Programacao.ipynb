## Dataset de Correção de Bugs

Este dataset foi utilizado na pesquisa publicada em [1] e consiste em um conjunto de informações sobre tarefas de correção de bugs no projeto [Apache Hadoop](https://hadoop.apache.org/), coletadas ao longo de um período de `10` anos. Além disso, o dataset foi manualmente rotulado para indicar se o bug registrado é ou não um bug de tratamento de exceção.

```bash
[1] da Silva, A.J.A., Vieira, R.G., Mesquita, D.P.P. et al. Towards automatic labeling of exception handling bugs: A case study of 10 years bug-fixing in Apache Hadoop. Empir Software Eng 29, 85 (2024). https://doi.org/10.1007/s10664-024-10494-0
```

A arquitetura do Hadoop compreende quatro componentes principais:

- `HADOOP` (ou core): que fornece o pacote de utilitários para suportar outros módulos do Hadoop;
- `MAPREDUCE`: um modelo de programação para armazenamento e processamento de dados;
- `HDFS` (*Hadoop Distributed File System*): um sistema de arquivos distribuído que roda em clusters projetados para armazenar arquivos muito grandes e fornecer acesso de alta taxa de transferência; e
- `YARN` (*Yet Another Resource Negotiator*): um framework para agendamento/monitoramento de tarefas e gerenciamento de recursos do cluster.

## Questões de Compreensão

**Questão 1** - Em que dia da semana mais bugs de tratamento de exceção são corrigidos?

````python
# Filtrar apenas os bugs de tratamento de exceção (Type = 1)
exception_bugs_df = df[df['Type'] == 1].copy()

# Converter a coluna 'ResolutionDate' para datetime
exception_bugs_df['ResolutionDate'] = pd.to_datetime(exception_bugs_df['ResolutionDate'], errors='coerce')

# Remover linhas onde a conversão de data falhou (NaN)
exception_bugs_df.dropna(subset=['ResolutionDate'], inplace=True)

# Extrair o dia da semana (0=Segunda, 6=Domingo)
exception_bugs_df['DayOfWeek'] = exception_bugs_df['ResolutionDate'].dt.dayofweek

# Mapear os números dos dias da semana para nomes para melhor legibilidade
day_names = {0: 'Segunda-feira', 1: 'Terça-feira', 2: 'Quarta-feira', 3: 'Quinta-feira', 4: 'Sexta-feira', 5: 'Sábado', 6: 'Domingo'}
exception_bugs_df['DayOfWeekName'] = exception_bugs_df['DayOfWeek'].map(day_names)

# Contar a frequência de bugs corrigidos por dia da semana
day_of_week_counts = exception_bugs_df['DayOfWeekName'].value_counts()

# Encontrar o dia da semana com mais correções
most_common_day = day_of_week_counts.idxmax()
max_bugs = day_of_week_counts.max()

print(f"O dia da semana em que mais bugs de tratamento de exceção são corrigidos é: {most_common_day} (com {max_bugs} bugs).")
````

**Questão 2** - No módulo `HDFS`, a maioria dos bugs de tratamento de exceção possuem qual prioridade?

````python
# Filtrar o DataFrame para o módulo HDFS e bugs de tratamento de exceção
hdfs_exception_bugs = df[(df['Project'] == 'HDFS') & (df['Type'] == 1)]

# Verificar se há bugs encontrados
if not hdfs_exception_bugs.empty:
    # Contar a frequência das prioridades
    priority_counts = hdfs_exception_bugs['Priority'].value_counts()

    # Encontrar a prioridade mais comum
    most_common_priority = priority_counts.idxmax()
    count_most_common = priority_counts.max()

    print(f"No módulo HDFS, a maioria dos bugs de tratamento de exceção possui a prioridade: '{most_common_priority}' (com {count_most_common} ocorrências).")
else:
    print("Não foram encontrados bugs de tratamento de exceção no módulo HDFS.")
````

**Questão 3** - Qual o módulo possui o menor tempo médio de correção de bug?

````python
# Converter as colunas de data para o tipo datetime
df['CreationDate'] = pd.to_datetime(df['CreationDate'], errors='coerce')
df['ResolutionDate'] = pd.to_datetime(df['ResolutionDate'], errors='coerce')

# Remover linhas onde a conversão de data falhou ou onde a ResolutionDate é anterior à CreationDate (bugs não resolvidos ou dados inconsistentes)
df_cleaned = df.dropna(subset=['CreationDate', 'ResolutionDate'])
df_cleaned = df_cleaned[df_cleaned['ResolutionDate'] >= df_cleaned['CreationDate']]

# Calcular o tempo de correção em dias
df_cleaned['CorrectionTime'] = (df_cleaned['ResolutionDate'] - df_cleaned['CreationDate']).dt.days

# Calcular o tempo médio de correção por módulo
mean_correction_time_per_project = df_cleaned.groupby('Project')['CorrectionTime'].mean()

# Encontrar o módulo com o menor tempo médio de correção
module_with_min_correction_time = mean_correction_time_per_project.idxmin()
min_correction_time = mean_correction_time_per_project.min()

print(f"O módulo com o menor tempo médio de correção de bug é '{module_with_min_correction_time}' com um tempo médio de {min_correction_time:.2f} dias.")
````

**Questão 4** - Qual o módulo possui o menor tempo médio de correção de bug de tratamento de exceção?

````python
# Filtrar apenas os bugs de tratamento de exceção (Type = 1)
exception_bugs_only_df = df[df['Type'] == 1].copy()

# Converter as colunas de data para o tipo datetime
exception_bugs_only_df['CreationDate'] = pd.to_datetime(exception_bugs_only_df['CreationDate'], errors='coerce')
exception_bugs_only_df['ResolutionDate'] = pd.to_datetime(exception_bugs_only_df['ResolutionDate'], errors='coerce')

# Remover linhas onde a conversão de data falhou ou onde a ResolutionDate é anterior à CreationDate
df_exception_cleaned = exception_bugs_only_df.dropna(subset=['CreationDate', 'ResolutionDate'])
df_exception_cleaned = df_exception_cleaned[df_exception_cleaned['ResolutionDate'] >= df_exception_cleaned['CreationDate']]

# Verificar se há dados após a limpeza
if not df_exception_cleaned.empty:
    # Calcular o tempo de correção em dias para bugs de exceção
    df_exception_cleaned['CorrectionTime'] = (df_exception_cleaned['ResolutionDate'] - df_exception_cleaned['CreationDate']).dt.days

    # Calcular o tempo médio de correção por módulo para bugs de exceção
    mean_correction_time_per_project_exception = df_exception_cleaned.groupby('Project')['CorrectionTime'].mean()

    # Encontrar o módulo com o menor tempo médio de correção de bugs de exceção
    module_with_min_correction_time_exception = mean_correction_time_per_project_exception.idxmin()
    min_correction_time_exception = mean_correction_time_per_project_exception.min()

    print(f"O módulo com o menor tempo médio de correção de bug de tratamento de exceção é '{module_with_min_correction_time_exception}' com um tempo médio de {min_correction_time_exception:.2f} dias.")
else:
    print("Não foram encontrados dados válidos de bugs de tratamento de exceção após a limpeza para calcular o tempo médio de correção.")
````

**Questão 5** -  O tempo médio de correção de bugs de tratamento de exceção é menor que o tempo médio de correção de outros bugs no módulo `MAPREDUCE`?

````python
# Filtrar o DataFrame para o módulo MAPREDUCE
mapreduce_df = df_cleaned[df_cleaned['Project'] == 'MAPREDUCE'].copy()

# Separar bugs de tratamento de exceção e outros bugs
exception_bugs_mapreduce = mapreduce_df[mapreduce_df['Type'] == 1]
other_bugs_mapreduce = mapreduce_df[mapreduce_df['Type'] == 0]

# Calcular o tempo médio de correção para bugs de tratamento de exceção no MAPREDUCE
mean_correction_time_exception = exception_bugs_mapreduce['CorrectionTime'].mean()

# Calcular o tempo médio de correção para outros bugs no MAPREDUCE
mean_correction_time_other = other_bugs_mapreduce['CorrectionTime'].mean()

print(f"No módulo MAPREDUCE:")
print(f"  - Tempo médio de correção para bugs de tratamento de exceção: {mean_correction_time_exception:.2f} dias")
print(f"  - Tempo médio de correção para outros bugs: {mean_correction_time_other:.2f} dias")

if mean_correction_time_exception < mean_correction_time_other:
    print(f"O tempo médio de correção de bugs de tratamento de exceção é MENOR que o tempo médio de correção de outros bugs no módulo MAPREDUCE.")
else:
    print(f"O tempo médio de correção de bugs de tratamento de exceção é IGUAL ou MAIOR que o tempo médio de correção de outros bugs no módulo MAPREDUCE.")
````
