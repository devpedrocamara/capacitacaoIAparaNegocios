> Obs: Crie a variável `ResolutionTime` e adicione ao dataframe, como o tempo total para solução do bug em dias como sendo `ResolutionDate` - `CreationDate`.

````python
import pandas as pd

# URL do dataset
url = "https://raw.githubusercontent.com/emanueles/EAN/refs/heads/main/labeled-hadoop-bug-fix-dataset.csv"

# Carregar o dataset
df = pd.read_csv(url, sep=';')

# Exibir as primeiras linhas do DataFrame para verificar o carregamento
df.head()
````

````python
df['CreationDate'] = pd.to_datetime(df['CreationDate'])
df['ResolutionDate'] = pd.to_datetime(df['ResolutionDate'])
df['ResolutionTime'] = (df['ResolutionDate'] - df['CreationDate']).dt.days

print("First 5 rows with new 'ResolutionTime' column:")
print(df[['CreationDate', 'ResolutionDate', 'ResolutionTime']].head())
````

````python
### Resposta da questão 1

import numpy as np

# Lista de variáveis numéricas para as quais as estatísticas descritivas serão calculadas
numerical_vars = [
    'NoComments', 'NoCommits', 'NoAttachments', 'NoAttachedPatches',
    'NoAuthors', 'SrcAddLines', 'SrcDelLines', 'ResolutionTime'
]

# Cria um DataFrame vazio para armazenar os resultados das estatísticas descritivas
descriptive_stats = pd.DataFrame(index=numerical_vars,
                                   columns=['Mean', 'Median', 'Std Dev', 'Coeff Var', 'Skewness', 'Kurtosis'])

# Itera sobre cada variável numérica
for col in numerical_vars:
    # Remove valores inválidos (NaNs) da série para evitar erros nos cálculos
    series = df[col].dropna()

    # Verifica se a série não está vazia após a remoção de NaNs
    if not series.empty:
        mean_val = series.mean()
        std_val = series.std()

        # Calcula e armazena as medidas de posição e dispersão
        descriptive_stats.loc[col, 'Mean'] = mean_val
        descriptive_stats.loc[col, 'Median'] = series.median()
        descriptive_stats.loc[col, 'Std Dev'] = std_val
        descriptive_stats.loc[col, 'Skewness'] = series.skew() # Assimetria
        descriptive_stats.loc[col, 'Kurtosis'] = series.kurtosis() # Curtose

        # Coeficiente de Variação (desvio padrão / média * 100)
        if mean_val != 0:
            descriptive_stats.loc[col, 'Coeff Var'] = (std_val / mean_val) * 100
        else:
            # Se a média for zero, o coeficiente de variação é indefinido
            descriptive_stats.loc[col, 'Coeff Var'] = np.nan
    else:
        # Se todos os valores forem inválidos para a coluna, preenche com NaN
        descriptive_stats.loc[col] = np.nan

# Exibe o DataFrame com as estatísticas descritivas calculadas
print("Descriptive Statistics for Numerical Variables:")
display(descriptive_stats)
````

### Questão 3: Detecção de outliers

Descubra se existem e quais são os outliers de : **NoComments, NoAttachments, NoAttachedPatches e ResolutionTime**, usando a amplitude interquartil (IQR). Plote os boxplots de cada variável.

````python
### Resposta da questão 3

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Variáveis para as quais os outliers serão analisados
outlier_vars = ['NoComments', 'NoAttachments', 'NoAttachedPatches', 'ResolutionTime']
outliers_data = {}

# Itera sobre cada variável para análise de outliers e plotagem de boxplot
for col in outlier_vars:
    # Remove valores NaN para o cálculo
    series = df[col].dropna()

    # Verifica se a série não está vazia após a remoção de NaNs
    if not series.empty:
        # Calcula Q1 (25º percentil) e Q3 (75º percentil)
        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1 # Amplitude Interquartil

        # Calcula os limites para identificação de outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identifica os outliers
        outliers = series[(series < lower_bound) | (series > upper_bound)]

        # Armazena os dados dos outliers
        outliers_data[col] = {
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'Lower Bound': lower_bound,
            'Upper Bound': upper_bound,
            'Outliers Count': len(outliers),
            'Outliers Values': outliers.values.tolist() # Converte para lista para exibição
        }

        print(f"\n--- Análise de Outliers para: {col} ---")
        print(f"Q1: {Q1:.2f}")
        print(f"Q3: {Q3:.2f}")
        print(f"IQR: {IQR:.2f}")
        print(f"Limite Inferior (Lower Bound): {lower_bound:.2f}")
        print(f"Limite Superior (Upper Bound): {upper_bound:.2f}")
        print(f"Número de Outliers: {len(outliers)}")
        if len(outliers) > 0:
            print(f"Valores dos Outliers (primeiros 10): {outliers.head(10).tolist()}...")
        else:
            print("Nenhum outlier detectado.")

        # Plota o Boxplot na escala normal
        plt.figure(figsize=(10, 6))
        sns.boxplot(y=series) # Usa 'y' para um boxplot vertical
        plt.title(f'Boxplot de {col} (Escala Normal)')
        plt.ylabel(col)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()

        # Plota o Boxplot com escala logarítmica para melhorar a visualização
        plt.figure(figsize=(10, 6))
        sns.boxplot(y=series) # Usa 'y' para um boxplot vertical
        plt.title(f'Boxplot de {col} (Escala Logarítmica no Eixo Y)')
        plt.ylabel(col)
        plt.yscale('log') # Aplica escala logarítmica
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()
    else:
        print(f"\n--- Análise de Outliers para: {col} ---")
        print("A série está vazia após a remoção de NaNs. Não é possível calcular outliers.")
````

### Questão 4: Intervalo de Confiança sem Bootstrap

Calcule os intervalos de confiança 90%, 95% e 99% para a média de tempo de resolução de bug (ResolutionTime), supondo que a distribuição amostral da média de ResolutionTime seja normal.

````python
### Resposta da questão 4

from scipy import stats

# Remove valores NaN da coluna ResolutionTime para garantir cálculos corretos
resolution_time_data = df['ResolutionTime'].dropna()

# Calcula a média, desvio padrão e o número de observações da amostra
mean_resolution_time = resolution_time_data.mean()
std_resolution_time = resolution_time_data.std()
n_resolution_time = resolution_time_data.count()

# Calcula o Erro Padrão da Média (Standard Error of the Mean - SEM)
sem_resolution_time = std_resolution_time / np.sqrt(n_resolution_time)

# Define os níveis de confiança e calcula os intervalos
confidence_levels = [0.90, 0.95, 0.99]

print(f"Média de ResolutionTime: {mean_resolution_time:.2f} dias")
print(f"Desvio Padrão de ResolutionTime: {std_resolution_time:.2f} dias")
print(f"Número de Observações (n): {n_resolution_time}")
print(f"Erro Padrão da Média (SEM): {sem_resolution_time:.2f} dias\n")

for confidence_level in confidence_levels:
    # Calcula o valor crítico Z para o nível de confiança
    # (1 - confidence_level) / 2 para obter a área da cauda
    z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Calcula a margem de erro
    margin_of_error = z_critical * sem_resolution_time

    # Calcula o intervalo de confiança
    lower_bound = mean_resolution_time - margin_of_error
    upper_bound = mean_resolution_time + margin_of_error

    print(f"Intervalo de Confiança de {confidence_level*100:.0f}%:")
    print(f"  Z-crítico: {z_critical:.3f}")
    print(f"  Margem de Erro: {margin_of_error:.2f} dias")
    print(f"  [{lower_bound:.2f}, {upper_bound:.2f}] dias\n")
````

### Questão 6: Regressão Linear Simples

Será que é possível prever o tempo de solução de bug (ResolutionTime) a partir do número de comentários (NoComments) da issue?

Plote o scatterplot das variáveis, depois construa o modelo de regressão linear simples usando o método dos mínimos quadrados e plote os gráficos de resíduos para responder à questão 6 do formulário.

````python
### Resposta da questão 6

# Cria um DataFrame limpo removendo linhas com NaN em 'NoComments' ou 'ResolutionTime'
df_cleaned = df.dropna(subset=['NoComments', 'ResolutionTime'])

# --- Primeiro Scatterplot: Escalas Normais ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NoComments', y='ResolutionTime', data=df_cleaned)
plt.title('Gráfico de Dispersão de ResolutionTime vs. NoComments')
plt.xlabel('NoComments')
plt.ylabel('ResolutionTime (dias)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --- Segundo Scatterplot: Eixo X Logarítmico ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x='NoComments', y='ResolutionTime', data=df_cleaned)
plt.title('Gráfico de Dispersão de ResolutionTime vs. NoComments (Escala Logarítmica para Eixo X)')
plt.xlabel('NoComments (escala logarítmica)')
plt.ylabel('ResolutionTime (dias)')
plt.xscale('log') # Aplica escala logarítmica ao eixo x
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

print()

import statsmodels.formula.api as smf

# Define a fórmula da regressão
# Variável dependente: ResolutionTime
# Variável independente: NoComments
formula = 'ResolutionTime ~ NoComments'

# Ajusta o modelo OLS usando o DataFrame df_cleaned
# df_cleaned garante que os valores NaN em 'NoComments' ou 'ResolutionTime' sejam tratados
model = smf.ols(formula, data=df_cleaned).fit()

# Imprime o resumo do modelo
print(model.summary())

# --- Plotagem dos Resíduos ---

# Gráfico de Resíduos vs. Valores Ajustados
plt.figure(figsize=(10, 6))
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.xlabel('Valores Ajustados (Fitted Values)')
plt.ylabel('Resíduos (Residuals)')
plt.title('Gráfico de Resíduos vs. Valores Ajustados')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Histograma dos Resíduos
plt.figure(figsize=(10, 6))
sns.histplot(model.resid, kde=True)
plt.xlabel('Resíduos')
plt.ylabel('Frequência')
plt.title('Histograma dos Resíduos')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Q-Q plot dos Resíduos para verificar a normalidade
import statsmodels.api as sm
plt.figure(figsize=(10, 6))
sm.qqplot(model.resid, line='s', ax=plt.gca())
plt.title('Q-Q Plot dos Resíduos')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

### Questão 7: Comparação entre dois grupos

Divida o dataset entre dois grupos: o grupo em que o tipo de bug é tratamento de exceção (IsExceptionBug='ExceptionBug') e o grupo formado pelos demais tipos de bug.

Conduza um teste de hipótese com nível de significância de 0.05 para a diferença das médias de ResolutionTime. Pressuponha que a distribuição da diferença das médias dos dois grupos segue uma distribuição normal. Calcule o tamanho do efeito do teste usando o d de Cohen.

````python
### Resposta da questão 7

from scipy.stats import ttest_ind

# Filtra para bugs de exceção (Type == 1) e extrai 'ResolutionTime', removendo NaNs
exception_bugs = df[df['Type'] == 1]['ResolutionTime'].dropna()

# Filtra para outros bugs (Type == 0) e extrai 'ResolutionTime', removendo NaNs
other_bugs = df[df['Type'] == 0]['ResolutionTime'].dropna()

print("--- Bugs de Exceção (ResolutionTime) ---")
print("Primeiras 5 entradas:")
print(exception_bugs.head())
print(f"Total de entradas não nulas: {len(exception_bugs)}\n")

print("--- Outros Bugs (ResolutionTime) ---")
print("Primeiras 5 entradas:")
print(other_bugs.head())
print(f"Total de entradas não nulas: {len(other_bugs)}")

# Realiza o teste t de amostras independentes (teste t de Welch, assumindo variâncias desiguais)
t_statistic, p_value = ttest_ind(exception_bugs, other_bugs, equal_var=False)

# Define o nível de significância
alpha = 0.05

print("--- Resultados do Teste t de Amostras Independentes ---")
print(f"Estatística T: {t_statistic:.3f}")
print(f"P-valor: {p_value:.3f}")

# Conclusão baseada no P-valor
if p_value < alpha:
    print(f"\nConclusão: Rejeitar a hipótese nula com alfa = {alpha}.\n")
    print("Isso sugere que há uma diferença estatisticamente significativa no ResolutionTime médio entre bugs de exceção e outros bugs.")
else:
    print(f"\nConclusão: Falha em rejeitar a hipótese nula com alfa = {alpha}.\n")
    print("Isso sugere que não há uma diferença estatisticamente significativa no ResolutionTime médio entre bugs de exceção e outros bugs.")

# Calcula o d de Cohen
def cohen_d(x, y):
    nx = len(x)
    ny = len(y)
    dof = nx + ny - 2
    # Usa ddof=1 para desvio padrão da amostra
    pooled_std = np.sqrt(((nx - 1) * np.std(x, ddof=1)**2 + (ny - 1) * np.std(y, ddof=1)**2) / dof)
    return (np.mean(x) - np.mean(y)) / pooled_std

cohens_d_value = cohen_d(exception_bugs, other_bugs)

print(f"\nTamanho do efeito d de Cohen: {cohens_d_value:.3f}")

# Interpreta o d de Cohen
if cohens_d_value < 0.2:
    interpretation = "muito pequeno"
elif cohens_d_value < 0.5:
    interpretation = "pequeno"
elif cohens_d_value < 0.8:
    interpretation = "médio"
else:
    interpretation = "grande"

print(f"Interpretação: O tamanho do efeito é {interpretation}, indicando uma diferença {interpretation} no ResolutionTime entre bugs de exceção e outros bugs.")
````

### Questão 8: Comparação entre dois grupos (parte 2)

Divida o dataset entre dois grupos: um grupo com as issues do Projeto Hadoop e um grupo com as issues do Projeto Yarn.

Conduza um teste de hipótese com nível de significância de 0.05 para a diferença das médias de ResolutionTime. Pressuponha que a distribuição da diferença das médias dos dois grupos segue uma distribuição normal. Calcule o tamanho do efeito do teste usando o r.

````python
### Resposta da questão 8

hadoop_bugs = df[df['Project'] == 'HADOOP']['ResolutionTime'].dropna()
yarn_bugs = df[df['Project'] == 'YARN']['ResolutionTime'].dropna()

print("--- Project HADOOP (ResolutionTime) ---")
print("Primeiras 5 entradas:")
print(hadoop_bugs.head())
print(f"Total de entradas não nulas: {len(hadoop_bugs)}\n")

print("--- Project YARN (ResolutionTime) ---")
print("Primeiras 5 entradas:")
print(yarn_bugs.head())
print(f"Total de entradas não nulas: {len(yarn_bugs)}")

# Realiza o teste t de amostras independentes (teste t de Welch, assumindo variâncias desiguais)
t_statistic, p_value = ttest_ind(hadoop_bugs, yarn_bugs, equal_var=False)

# Define o nível de significância
alpha = 0.05

print("--- Resultados do Teste t de Amostras Independentes (HADOOP vs. YARN) ---")
print(f"Estatística T: {t_statistic:.3f}")
print(f"P-valor: {p_value:.3f}")

# Conclusão baseada no P-valor
if p_value < alpha:
    print(f"\nConclusão: Rejeitar a hipótese nula com alfa = {alpha}.\n")
    print("Isso sugere que há uma diferença estatisticamente significativa no ResolutionTime médio entre bugs do Projeto HADOOP e do Projeto YARN.")
else:
    print(f"\nConclusão: Falha em rejeitar a hipótese nula com alfa = {alpha}.\n")
    print("Isso sugere que não há uma diferença estatisticamente significativa no ResolutionTime médio entre bugs do Projeto HADOOP e do Projeto YARN.")

# Calcula o r de Pearson como medida de tamanho de efeito para teste t
def pearson_r_from_t(t, df1, df2):
    # A tarefa pede explicitamente 'r', que pode ser derivado da estatística t para dois grupos independentes
    # A fórmula para r de Pearson a partir do t-statistic é r = t / sqrt(t^2 + df_total)
    # df_total são os graus de liberdade para o teste t de amostras independentes
    df_total = df1 + df2 - 2
    r = t / np.sqrt(t**2 + df_total)
    return r

# Obtém os graus de liberdade para cada grupo
df_hadoop = len(hadoop_bugs) - 1
df_yarn = len(yarn_bugs) - 1

# Calcula o r de Pearson para o tamanho do efeito
# Nota: para o teste t de amostras independentes, r é equivalente ao coeficiente de correlação ponto-bisserial
r_value = pearson_r_from_t(t_statistic, len(hadoop_bugs), len(yarn_bugs))

print(f"\nTamanho do efeito r de Pearson: {r_value:.3f}")

# Interpretação do r de Pearson
abs_r_value = abs(r_value)
if abs_r_value < 0.1:
    interpretation = "muito pequeno"
elif abs_r_value < 0.3:
    interpretation = "pequeno"
elif abs_r_value < 0.5:
    interpretation = "médio"
else:
    interpretation = "grande"

print(f"Interpretação: O tamanho do efeito é {interpretation}, indicando uma associação {interpretation} entre o projeto e o ResolutionTime.")
````
